```python
### add time range
years= ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']
start_month ='1'
end_month = '12'
```

## Construct URL and wrangle json to values dataframe by year and weather attribute, save data

```python
result = pd.DataFrame()
for year in years:
    df_ctx = w.get_data(year, start_month, end_month, el_kood, stations_l, inajar, tyyp, outdir)
    print(df_ctx.shape)
    print('nan or 0 values ....')
    df_ctx.isnull().values.sum()
    df_ctx = df_ctx.rename(columns={"Tallinn-Harku": "air_temp"})
    result=pd.concat([result, df_ctx], axis=0)

filename='air_temp_' + str(year) + '.txt'
result.to_csv(OUTPUT_DIRECTORY / filename)
```

### N-year average for one station



```python
start_year = 2014
end_year = 2025
year_range = end_year - start_year + 1
filename_prx = 'climate_file_'
filename_ext = '.txt'
df_kokku = pd.DataFrame(index=range(1,366))   ###to accumulate results!!!



for y in range(start_year, end_year):
    # print(str(y))
    filename = outdir + filename_prx + str(y) + filename_ext

    new_columns = []
    for kood in el_kood_l:
        coln = kood + '_' + str(y)
        new_columns.append(coln)
    #print(new_columns)

    df = pd.read_csv(filename, index_col = 0, header = 0)
    df.columns = new_columns
    df_kokku=pd.concat([df_kokku, df], axis=1)

#df_kokku
#new_columns
```

```python
# find average
df_average = pd.DataFrame(index=range(1,366))   ###to accumulate results!!!

for kood in el_kood_l:
    select_col = []
    for y in range(start_year, end_year):
        select_col.append(kood + '_' + str(y))
    df_average = pd.concat([df_average, df_kokku[select_col].mean(axis = 1)], axis=1)
df_average.columns = el_kood_l
df_average.index.name = 'DOY'
```
